```python
import pandas as pd

# Load the DataFrame
df = pd.read_csv("beliefs_emotion_scores.csv")
df.head()

# List of emotions (assuming high_income_dict.keys() contains the emotions)
emotions = list(high_income_dict.keys())

# List of income categories
income_categories = ["high_income", "upper_middle_income", "lower_middle_income", "low_income"]

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=["Emotion"] + income_categories)

# Iterate over each emotion
for emotion in emotions:
    # Create a list to store the mean values for this emotion across different income categories
    mean_values = [emotion]  # Start with the emotion name

    # Calculate mean for each income category
    for income_cat in income_categories:
        column_name = f"{income_cat}_{emotion}"  # Construct the column name
        mean_value = df[column_name].mean()  # Calculate the mean value for that column
        mean_values.append(mean_value)  # Add the mean to the list

    # Add the mean values as a new row to the DataFrame
    result_df.loc[len(result_df)] = mean_values

# Display the final DataFrame
result_df.to_csv("emotions_stats.csv", index=False)
files.download("emotions_stats.csv")

```
